```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import geopy.distance
import datetime
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('sample.csv')
df.head()
```

## Distancia do Centro 

```python
Centerdf = df.copy()
Centerdf = Centerdf.dropna(subset=['Location'])
```

```python
def distance_from_center(point):
    center = (41.8776142,-87.6323187)
    point = point.replace("(", "").replace(")", "")
    point = (point.split(",")[0], point.split(",")[1])
    return int(geopy.distance.distance(center, point).km)

Centerdf["Distance"] = Centerdf["Location"].apply(distance_from_center)
```

```python
distances = Centerdf.groupby(["Distance"]).size()
plt.figure(figsize=(16,10))
plt.plot(distances)
plt.xlabel('Distance in KM')
plt.ylabel('Number of reported crimes')
plt.title("Number of reported crime by distance from center")
plt.ylim(ymin=0)
plt.show()
```

## DiferenÃ§a entre regiÃ£o central e nÃ£o central

```python
def permuta(df1, df2, sample_len=500, number_perm=10000):
    temp = []
    for i in range(sample_len):
        j = random.randint(0, len(df1))
        temp.append(df1.iloc[j])
    df1 = np.array(temp)
    temp = []
    for i in range(sample_len):
        j = random.randint(0, len(df2))
        temp.append(df2.iloc[j])
    df2 = np.array(temp)
    
    diffs = []
    for _ in range(number_perm):
        df1_, df2_ = df1, df2
        j = random.randint(0, sample_len-1)
        temp = df1_[j]
        df1_[j] = df2_[j]
        df2_[j] = temp
        diff = df1_.sum() - df2_.sum()
        diffs.append(diff)
    return df1.sum() - df2.sum(), diffs
```

```python
block1 = Centerdf[Centerdf['Distance'] > 2]['Arrest']
block2 = Centerdf[Centerdf['Distance'] <= 2]['Arrest']
realDiff, diffs = permuta(block1, block2)
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,8))
plt.xlabel('Difference')
plt.hist(diffs, bins=20, color='royalblue', edgecolor='black', alpha=0.7)
plt.title('Differences between # prisons in downtown and non downtown', pad=40)
plt.axvline(realDiff, color='red', label='Real difference')
plt.legend(fontsize=15)
plt.plot()
Centerdf = 1
```

## Crimes por ano

```python
yeardf = df.copy()
yeardf['Datetime'] =  pd.to_datetime(df['Date'])
```

```python
yeardf["Week"] = yeardf['Datetime'].dt.week
yeardf["Year"] = yeardf['Datetime'].dt.year
yeardf = yeardf[yeardf["Week"] != 53]
occurences = yeardf.groupby(["Year", "Week"]).size()
plt.figure(figsize=(16,10))
plt.plot(list(occurences))
plt.xlabel('Week since 2015')
plt.ylabel('Number of reported crimes')
plt.title("Number of reported crime by week since 2015")
plt.ylim(ymin=0)
plt.show()
```

#### RegreessÃ£o do nÃºmero de crimes por semana nos anos reportados

```python
from sklearn.linear_model import LinearRegression
lis = list(occurences)
X = []
y = []
for a, b in enumerate(lis):
    X.append(a)
    y.append(b)
    
X = np.array(X).reshape(-1,1)
y = np.array(y).reshape(-1,1)

regressor = LinearRegression()
regressor.fit(X, y)
plt.figure(figsize=(16,10))
plt.plot(X,regressor.predict(X))
plt.plot(X,list(occurences))
plt.plot(list(occurences))
plt.xlabel('Week since 2015')
plt.ylabel('Number of reported crimes')
plt.title("Number of reported crime by week since 2015 - Regression")
plt.ylim(ymin=0)
plt.show()
yeardf = 1
```
