```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
df = pd.read_csv('Crimes_2015_to_2018.csv')
```

```python
df.head()
```

```python
domestic = df[df['Domestic'] == True]['Arrest']
nonDomestic = df[df['Domestic'] == False]['Arrest']
realDomesticArrest = domestic.sum() # Number of arrests in a domestic environment 
realNonDomesticArrest = nonDomestic.sum() # Number of arrests in a non domestic environment 
realDiff = abs(realDomesticArrest - realNonDomesticArrest) # Difference of arrests in domestic and non domestic environments

numPerm = 10000
diffs, domesticArrest, nonDomesticArrest = [], [], []
originalDomestic = df['Domestic'].copy()
```

```python
# Permuatation
for _ in np.arange(numPerm):
    df['Domestic'] = df['Domestic'].sample(frac=1).reset_index(drop=True)
    
    filter = df['Domestic'] == True
    domestic = df[filter]['Arrest']
    nonDomestic = df[~filter]['Arrest']
    domesticArrest.append(domestic.sum())
    nonDomesticArrest.append(nonDomestic.sum())
    diffs.append(abs(domesticArrest[-1] - nonDomesticArrest[-1]))
```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,8))
plt.xlabel('Difference')
plt.hist(diffs, bins=20, color='royalblue', edgecolor='black', alpha=0.7)
plt.title('Differences between # prisons in domestic and non domestic environments', pad=40)
plt.axvline(realDiff, color='red', label='Real difference')
plt.legend(fontsize=15)
```
